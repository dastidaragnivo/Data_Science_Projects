```python
#Problem Statement 1: Samantha has created a dataset named 'top50spotify.csv' of her top 50 songs from spotify.

"""
Dataset Description:top50spotify.csv -
The dataset contains 14 features. 
Here's a brief description of a few columns in the dataset:
•SerialNo. -Serial number of songs
•Track.Name -Name of the track
•Artist.Name -Name of the artist
•Genre -Genre of the song
•Energy -Energy index of the song
•Length. -Length of the song
•Popularity -Popularity index of the song

Tasks to be performed:
1.Import the dataset as a DataFrame and drop the first column.
2.Save it as 'top50.csv'.
3.Find the average Energy and Length of first 10 songs.
4.Find the total length of songs, group by genre from top to bottom.
5.Print the artist name with the most number of tracks in one genre. 
(Hint: Group by artist name and genre)
6.Print the data of the tracks created by the artist from the previous question.
"""
```

```python
#Task 1.Import the dataset as a DataFrame and drop the first column

import pandas as pd
spotify_df=pd.read_csv("top50spotify.csv")
spotify_df.drop(["SerialNo."],axis=1,inplace=True)
spotify_df
```

```python
#Task 2.Save it as 'top50.csv'

spotify_df.to_csv("top50.csv")
```

```python
#Task 3.Find the average Energy and Length of first 10 songs

#Average Energy of 1st 10 songs
print((spotify_df.head(10))["Energy"].agg("mean"))

#Average Length of 1st 10 songs
print((spotify_df.head(10))["Length."].agg("mean"))
```

```python
#Task 4.Find the total length of songs, group by genre from top to bottom

print(spotify_df.groupby("Genre")["Length."].agg("sum"))
```

```python
#Task 5.Print the artist name with the most number of tracks in one genre

print((spotify_df.groupby(["Artist.Name","Genre"])["Track.Name"].agg("count").sort_values(ascending=False)).head(1))
```

```python
#Task 6.Print the data of the tracks created by the artist from the previous question

print(spotify_df["Track.Name"].loc[spotify_df["Artist.Name"]=="Ed Sheeran"])
```

```python
#Problem Statement 2: Write a Python program to perform the following tasks-

"""
1.Create a pandas series from the below dictionary where indices are subjects:
{'English':{'Sam':60,'Jackson':74,'Ahree':85}, 'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}}
2.Convert the created series into DataFrame and replace the null values with zeroes.
3.Transpose the DataFrame and create a new column 'Average' and fill the values in it by calculating the average of all subjects.
"""
```

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

marks_series=pd.Series({'English':{'Sam':60,'Jackson':74,'Ahree':85}, 
                        'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},
                        'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},
                        'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},
                        'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}})
print(marks_series)
std_dict={}
for key in ['Ahree','Aron','Gloria','Gray','Isla','Jackson','Sam']:
    std_dict[key]=[]

print("Initial dictionary:",std_dict)

#Convert series.values numpy ndarray to dictionary
for arr in marks_series.values:
    std_series=pd.Series(arr,index=['Ahree','Aron','Gloria','Gray','Isla','Jackson','Sam'])
    for i in range(len(std_series)):
        std_dict[std_series.index[i]].append(std_series[i])


print("Updated dictionary:",std_dict)

subject_df=pd.DataFrame(marks_series.index)
print(subject_df)

std_df=pd.DataFrame(std_dict)
print(std_df)

#Combined Dataframe
std_marks_df=pd.concat([subject_df,std_df],axis=1)
print("Dataframe with null values:")
print(std_marks_df)

#Replace null values with zero
std_marks_df=std_marks_df.fillna(0)
print("Dataframe without null values:")
print(std_marks_df)

#Tranpose Dataframe
std_marks_df_trans=std_marks_df.T
print(std_marks_df_trans)

#Add new column "Average" to the new dataframe
#Before that set the first row as header
std_marks_df_trans.columns=std_marks_df_trans.iloc[0]
std_marks_df_trans = std_marks_df_trans[1:]
std_marks_df_trans['Average']=std_marks_df_trans.mean(axis=1)
print(std_marks_df_trans)

```

```python
#Problem Statement 3: Write a Python program to create a series from 1 to 1000 and select only numbers divisible by 7 and 17.
```

```python
import numpy as np
import pandas as pd

num_series=pd.Series(np.arange(1,1000,1))
result_set=[x for x in num_series.values if (x%7)==0 and (x%17)==0]
print("Answer1:",result_set)

#Alternatively

num=[]
for i in range(1,1001):
    if (i%7)==0 and (i%17)==0:
        num.append(i)

print("Answer2:",num)
```

```python
#Problem Statement 4: Sylphia has a dataset of various cereals sold in the supermarket.
"""
Dataset Description:cereal.csv-The dataset contains 16 features. 

Here's a brief description of 3 columns in the dataset:
•name-Brand name of the cereals
•MFR-Manufacturer of the brands
•rating-Rating of the cereals

Syliphia wants to visualize the quality of cereals and determine which manufacturer delivers the best quality.
Tasks to be performed:
1.Import the dataset.
2.Plot ratings of different types of manufacturers.
3.Use xticks range form 0-100.
4.Change the style of the graph to seaborn
"""
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Import the csv data in a dataframe
cereal_df=pd.read_csv(r"cereal.csv")
print(cereal_df)

#Plot ratings of different types of manufacturers
plt.figure(figsize=(30,20))
plt.plot(cereal_df["rating"])
plt.xticks(np.arange(0,100))
plt.ylabel("Rating per manufacturer",fontsize=20)
plt.title("Rating plot for different manufacturers",fontweight="bold", fontsize=16)


#Change graph style to seaborn
sns.set()
sns.set_palette("deep")
plt.xticks(np.arange(0,100))
plt.ylabel("Rating per manufacturer",fontsize=20)
plt.title("Rating plot for different manufacturers",fontweight="bold", fontsize=16)
sns.set_style("darkgrid")
plt.plot(cereal_df["rating"])
plt.show()
```
