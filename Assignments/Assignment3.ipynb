```python
import pandas as pd
import scipy.stats as stats
import math
import numpy as np


X = int(input("Enter speed in km/hr:"))
print('The speed of the car checked at a given point of time is',X,'km/hr')

mean = 75
standard_deviation = 15

#Create normal distribution using 'norm' class
#loc -> specifies the mean
#scale -> specifies the standard deviation

speed_distribution = stats.norm(loc=mean,scale=standard_deviation)

#Find probability that a car picked at random is traveling at more than 'X' km/hr
#Use cdf() method from norm class. CDF -> Cumulative Density Function. CDF gives you the probability of a normally distributed population that is less than or equal to a value.
#Subtract CDF output from 1 to get probability of a normally distributed population that is more than the value

probability_Xg = 1 - speed_distribution.cdf(x=X)
print("Probability that the car is traveling at more than X km/hr =",probability_Xg.round(4))

```

```python
flood_df_500 = flood_df[["YEAR","JUN","JUL"]]
flood_df_500.head()
```

```python
flood_df_500["JUN_GT_500"] = np.where(flood_df_500["JUN"]>500,1,0)
flood_df_500["JUL_GT_500"] = np.where(flood_df_500["JUL"]>500,1,0)
flood_df_500["FLOODS"] = np.where(flood_df["FLOODS"]=="YES",1,0)
flood_df_500.drop(["JUN","JUL"],axis=1,inplace=True)
flood_df_500.head()
```

```python
print("Probability of flood given it rained more than 500 mm in June (P(A|B)):")

# define the event A (flood occuring in a year)
event_A = flood_df_500.loc[np.where(flood_df_500["FLOODS"] == 1)]

# define the event B (rainfall greater than 500 mm in June)
event_B = flood_df_500.loc[np.where(flood_df_500["JUN_GT_500"] == 1)]

# calculate P(A ∩ B)
event_A_intersect_B = flood_df_500.loc[np.where((flood_df_500["FLOODS"] == 1) & (flood_df_500["JUN_GT_500"] == 1))]
prob_A_intersect_B = event_A_intersect_B["YEAR"].count() / 118

# calculate P(B)
prob_B = event_B["YEAR"].count() / 118

# calculate P(A|B)
prob_A_given_B = prob_A_intersect_B / prob_B
print(prob_A_given_B.round(5))
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import math
import seaborn as sns
import sklearn
from sklearn.datasets import load_wine

wine_bunch = load_wine()

print(wine_bunch.keys())

print(wine_bunch.DESCR)
```

```python
#Convert the dataset into DataFrame using pandas
wine_df = pd.DataFrame(wine_bunch.data,columns=wine_bunch.feature_names)
wine_df["Class"]=wine_bunch.target
wine_df
```

```python
wine_df["Class"].value_counts()
```

```python
wine_df.isnull().sum()
```

```python
#Generate the sample size of 50 and give a random state as 100

sample_size = 50
wine_data_sample = wine_df.sample(n=50, random_state=100)
wine_data_sample
```

```python
#Calculate Z-critical, Margin of Error,and Confidence Interval for "alcohol" at 95% significance interval on the generated sample data

sample_mean = wine_data_sample["alcohol"].mean()
sample_std = wine_data_sample["alcohol"].std()

a = math.sqrt(sample_size)
z_critical = stats.norm.ppf(q=0.95)
print("Z-critical value:",z_critical)

#M.O.E = z-score * (S.D / sample size**(1/2))
margin_of_error = z_critical * (sample_std/a)
print("Margin of Error:",margin_of_error)

#Confidence Interval = (Sample Mean + M.O.E, Sample Mean - M.O.E)
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
print("Confidence Interval:",confidence_interval)

```
