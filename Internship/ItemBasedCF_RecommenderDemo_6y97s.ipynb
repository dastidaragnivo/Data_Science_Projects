```python
import pandas as pd
```

```python
ratings_df = pd.read_csv('D:/Edureka/DS&ML_internship/ml-100k/ml-100k/u.data' , sep ='\t' ,
                      names=['user_id','item_id','rating' , 'timestamp'])

movies_df = pd.read_csv('D:/Edureka/DS&ML_internship/ml-100k/ml-100k//u.item', sep='|', 
                     names=['movie_id','movie_title','release_date','video_release_date',
                            'IMDb URL','unknown','Action','Adventure','Animation','Children',
                            'Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror',
                            'Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western'], 
                     encoding='latin-1')

movies_df = movies_df.iloc[:,[0,1]]
```

# Let's build the association matrix

```python
learningMatrix = ratings_df.pivot_table(index=["item_id"], columns=["user_id"],
                                        values="rating").reset_index(drop=True)

learningMatrix.head(10)
```

```python
learningMatrix.shape
```

```python
learningMatrix.fillna(0,inplace = True)
learningMatrix.head(20)
```

# Find out the similarity between the items

```python
from sklearn.metrics import pairwise_distances
import numpy as np
```

```python
movie_similarity = pairwise_distances( learningMatrix, metric="cosine" )
np.fill_diagonal( movie_similarity, 0 )
ratings_matrix = pd.DataFrame(movie_similarity)

ratings_matrix.head(10)
```

```python
ratings_matrix.shape
```

```python
try:
    user_inp=input('Enter the reference movie title based on which recommendations are to be made: ')
    inp=movies_df[movies_df['movie_title']==user_inp].index.tolist()
    
    inp=inp[0]    
    

    movies_df['similarity'] = ratings_matrix.iloc[inp]
    #print(movies.head(5))
    print("Recommended movies based on your choice of ",user_inp ,": \n", movies_df.sort_values( ["similarity"], ascending = False )[1:10])
    
except:
    print("The movie name you have entered does not exist in the list, however, below are the top movies recommended in general")
    movies_df= movies_df.drop("similarity", axis=1)
    print(movies_df.head(10))
```
